```python
# Creating dictionary

def count_performance_rows(df):
    dictionary = {}
    for index, row in df.iterrows():
        loan_id = row['loan_id']
        if loan_id not in dictionary:
                    dictionary[loan_id] = {
                        'default_status': row['default_status'],
                        'loan_age': row['loan_age']
                    }
    return dictionary
#     print row['c1'], row['c2']
```

```python
dictionary = count_performance_rows(result)
dictionary
```

```python
def get_performance_summary_value(loan_id, key, dictionary):
    value = dictionary.get(loan_id, {
        "default_status": False,
        "loan_age": 0
    })
    return value[key]
```

#### So now, there are no missing values in the dataframe

```python
import numpy as np

def outliers_z_score(ys):
    threshold = 3

    mean_y = np.mean(ys)
    stdev_y = np.std(ys)
    z_scores = [(y - mean_y) / stdev_y for y in ys]
    return np.where(np.abs(z_scores) > threshold)
```

```python
outliers_z_score(pdf)
```
